```python
import re

# Cheat sheet from pythonprogramming.net
# 
# Identifiers:
# # \d = any number
# \D = anything but a number
# \s = space
# \S = anything but a space
# \w = any letter
# \W = anything but a letter
# . = any character, except for a new line
# \b = space around whole words
# \. = period. must use backslash, because . normally means any character.
# 
# Modifiers:
# # {1,3} = for digits, u expect 1-3 counts of digits, or "places"
# + = match 1 or more
# ? = match 0 or 1 repetitions.
# * = match 0 or MORE repetitions
# $ = matches at the end of string
# ^ = matches start of a string
# | = matches either/or. Example x|y = will match either x or y
# [] = range, or "variance"
# {x} = expect to see this amount of the preceding code.
# {x,y} = expect to see this x-y amounts of the precedng code
# 
# White Space Charts:
# # \n = new line
# \s = space
# \t = tab
# \e = escape
# \f = form feed
# \r = carriage return
# 
# Characters to REMEMBER TO ESCAPE IF USED!
# # . + * ? [ ] $ ^ ( ) { } | \
# 
# Brackets:
# # [] = quant[ia]tative = will find either quantitative, or quantatative.
# [a-z] = return any lowercase letter a-z
# [1-5a-qA-Z] = return all numbers 1-5, lowercase letters a-q and uppercase A-Z
# () = characters between to look for, i.e. <p>.*?</p> returns anything that is inside paragraph 
# and at the same time trims <p></p>
```

```python
sample_text = 'Jessica is 15 years old, and Daniel is 27 years old. Edward is 97 years old, and his grandfather, Oscar, is 102.' 

ages = re.findall(r'\d{1,3}', sample_text)
names = re.findall(r'[A-Z]\w*', sample_text)
first_two_letters = re.findall(r'[A-Z]\w?', sample_text)
string_end = re.findall(r'[a-z]$', sample_text)

removed_spaces = re.findall(r'\S*', sample_text)
for item in removed_spaces:
    if item == '':
        removed_spaces.remove(item)

print(names, ages)
print(first_two_letters)
print(removed_spaces)
print(string_end)
```

```python
import threading
from queue import Queue
import time

print_lock = threading.Lock()

def thread_function():
    while True:
        worker = q.get()
        sample_function(worker)
        q.task_done()

def sample_function(worker):
    time.sleep(0.5)
    with print_lock:
        print(threading.current_thread().name, worker)
    

q = Queue()
for x in range(10):
    t = threading.Thread(target=thread_function)
    t.daemon = True
    t.start()

start = time.time()
    
for worker in range(20):
    q.put(worker)
    
q.join()

end = time.time()
print('Whole process took %i' % (end - start))
```

```python
import threading
from queue import Queue
import socket

print_lock = threading.Lock()
server = 'hackthissite.org'

def port_scanner(port):
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    try:
        con = s.connect((server,port))
        with print_lock:
            print("Port", port, "is open")
        con.close()
    except:
        pass

def threader():
    while True:
        worker = q.get()
        port_scanner(worker)
        q.task_done()

q = Queue()

for i in range(30):
    t = threading.Thread(target=threader)
    t.daemon = True
    t.start()

x = 101
for worker in range(1,x,1):
    q.put(worker)
    
q.join()

print("All ports scanned.")
```
